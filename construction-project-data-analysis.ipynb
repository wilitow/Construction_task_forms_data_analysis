### 2.2 Data Cleaning 

* locate any missing values 
* check for duplicated items

```python
#number of missing data points per column
missing_values_count = pm_tasks_data.isnull().sum()
missing_values_count[:]
```

```python
# how many total missing values do we have?
total_cells = np.product(pm_tasks_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
((total_missing/total_cells) * 100).round(2)
```

Overall, the dataset has approximatelty 11.86% of missing data. 

The columns with missing values include Target, To Package, Association, Images, Comments, Documents, Priority, Cause and Task Group. With the exception of the Target column, the rest are all columns with string values. The Priority column has over 10,000 missing values. If we were to drop these rows our dataset would be reduced significantly. Let's review the data and find a methodical approach to fill the  missing values. 

```python
pm_tasks_data['Priority'].value_counts()
```

Behavioural Failure is the most applied in the Priority column. The tasks associated with this category relate to Safety. 

```python
pm_tasks_data.loc[(pm_tasks_data['Priority'].isnull()) & (pm_tasks_data['Task Group']=='Safety'), 'Priority']='Behavioural Failure'
```

```python
#check if change applied 
pm_tasks_data.loc[pm_tasks_data['Task Group'] == 'Safety']
```

```python
#the remaining rows can be filled with Other.
pm_tasks_data['Priority'] = pm_tasks_data['Priority'].fillna('Other')
pm_tasks_data['Priority']
```

```python
pm_tasks_data['Target'].value_counts()
```

```python
pm_tasks_data['Target'].unique()
```

Given that the value 44011.0 has been assigned the most, we will use this value to fill in the missing data.

```python
pm_tasks_data['Target'] = pm_tasks_data['Target'].fillna(44011.0)
pm_tasks_data['Target']
```

Now, let's take a look at the column named, Association. It is unclear what this column is for, however the values in this column include 'FormAnswer'. For this reason, it would seem that the data is missing because it hasn't been recorded. We will fill the missing values with 'Other'.

```python
pm_tasks_data['Association'].value_counts()
```

```python
pm_tasks_data['Association'] = pm_tasks_data['Association'].fillna('FormAnswer')
pm_tasks_data['Association']
```

The column named, Cause also contains some missing values. They seem to have not been recorded. Existing values in this column include 'Safety documentation', ' House keeping', or 'plant and equipment'. It's fair to say that the missing values have simply not been recorded and similarly, we could fill them with
'Other'.

```python
pm_tasks_data['Cause'].value_counts().head(5)
```

```python
pm_tasks_data['Cause'] = pm_tasks_data['Cause'].fillna('JPC - Safety - House Keeping ')
pm_tasks_data['Cause']
```

The column named, To Package indicates which trade contractor or department the task is associated with. By looking at the full dataset, it is clear that the missing values are for tasks associated with design. Therefore, it would be fair to say the data has not been recorded rather than it doesn't exist. We can fill the missing values with 'Design'.

```python
pm_tasks_data['To Package'].value_counts()
```

```python
pm_tasks_data['To Package'] = pm_tasks_data['To Package'].fillna('Design')
pm_tasks_data['To Package']
```

The columns headed 'Images', 'Comments' and 'Documents' contain Boolean data types. There's approx. 5% or less of missing data in these columns. Rather than assume what the value should be, we will fill the missing values with 'N/A'. 

```python
pm_tasks_data[['Images','Comments','Documents']] = pm_tasks_data[['Images','Comments','Documents']].fillna('N/A')
pm_tasks_data[['Images','Comments','Documents']]
```

Finally, the column named, Task Group has 50 missing values. This column details which sub-group the task is associated with. 

```python
pm_tasks_data['Task Group'].value_counts()
```

Given that Safety has been assigned the most in this column, we will fill any missing values with the sub-category, "Safety". 

```python
pm_tasks_data['Task Group'] = pm_tasks_data['Task Group'].fillna('Safety')
pm_tasks_data['Task Group']
```

Let's run the previous code again to check for missing values in the dataset.

```python
# number of missing data points per column
missing_values_count = pm_tasks_data.isnull().sum()

# look at the # of missing points in each column
missing_values_count[:]
```

```python
# how many total missing values do we have?
total_cells = np.product(pm_tasks_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

All missing values have now been dealt with. Let's consider if there are any duplicates in the dataset.

```python
#non-numerical data columns 
non_numeric_data = list(pm_tasks_data.select_dtypes(exclude=('int', 'float')).columns)
print(f'Сolumns without numeric data: {", ".join(non_numeric_data)}.')
print(len(non_numeric_data))
```

The following columns contain non-numeric data:- Ref, Status, Location, Description, Created, Type, To Package, Status Changed, Association, OverDue, Images, Comments, Documents, Priority, Cause, Report Status, Task Group. Due to the nature of the dataset, we can expect to find duplicate values in certain columns such as Status and Location. Let's take a closer look at the Description column to narrow down any duplicates. 

```python
print('Duplication Location:', 100*(len(pm_tasks_data['Location']) - len(pm_tasks_data['Location'].drop_duplicates()))/len(pm_tasks_data['Location']))
```

```python
print('Duplication Status:', 100*(len(pm_tasks_data['Status']) - len(pm_tasks_data['Status'].drop_duplicates()))/len(pm_tasks_data['Status']))
```

```python
print('Duplication Description:', 100*(len(pm_tasks_data['Description']) - len(pm_tasks_data['Description'].drop_duplicates()))/len(pm_tasks_data['Description']))
```

By locating the duplicated rows, it is clear that the Task Description relates to multiple packages. For example, items 16 and 17 refer to "no ladder tags" but this applies to both the Temp electrics package and Precast concrete package. ### 3.3 Average time taken to close out a task by project

```python
subset_project_tasks.dtypes
```

```python
close_out_time = subset_project_tasks[['Ref', 'Status', 'Location', 'Description', 'Created', 'Status Changed', 'project', 'Task Group']]
close_out_time.head()
```

```python
date_completed = close_out_time['Status Changed']
date_created = close_out_time['Created']

close_out_time['Time to complete task'] = date_completed - date_created
close_out_time
```

```python
avg_close_out_time=close_out_time.groupby(['project','Task Group'])[['Time to complete task']].mean().sort_values('Time to complete task', ascending=False).reset_index()
avg_close_out_time['Time to complete task']=avg_close_out_time['Time to complete task'].dt.days
avg_close_out_time
```

```python
avg_close_out_time['project'].unique()
```

```python
fig=px.bar(avg_close_out_time, x='project', y='Time to complete task', color='Task Group', title='Time taken to complete tasks by project and department')
#fig.layout = dict(xaxis=dict(type="category"))
fig.show()
```

```python
#central trend of time to compltete a task 
close_out_time.describe().T
```

```python
#On average, it took 11 days to close out a task, with the longest taking 598 days. There is a minimum of -304 days,
#which suggests there could be some sort of input error. 
```

```python
#select columns from pandas df to be used in a SQL database 
cols_to_use=pm_tasks_data[['Created', 'To Package', 'Status Changed', 'project', 'Task Group']]
cols_to_use
```

```python
DATABASE_FILEPATH='const_project_data.db'

if os.path.exists(DATABASE_FILEPATH):
    os.remove(DATABASE_FILEPATH)
    
conn = sqlite3.connect(DATABASE_FILEPATH)
cur = conn.cursor()
```

```python
cur.execute('''

CREATE TABLE IF NOT EXISTS project_tasks (

project INTEGER, 
Created DATE, 
StatusChanged DATE, 
ToPackage VARCHAR, 
TaskGroup VARCHAR)

 ''')
conn.commit()

df=cols_to_use
df.to_sql('project_tasks', conn, if_exists = 'replace', index = False)

cur.execute('''
SELECT * FROM project_tasks
''')
```
